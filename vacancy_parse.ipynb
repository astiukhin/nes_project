```python
vacancies = []
base_url = "https://career.habr.com/vacancies?page={}&type=all"
urls = [base_url.format(page) for page in range(1, 127)]
for i, url in zip(range(len(urls)), urls):
    print(f'{i} is in progress')
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, features = 'html.parser')
    for card in soup.find_all('div', class_ = 'vacancy-card__inner'):
        vacancy_date = card.find('div', class_ = 'vacancy-card__date').text
        vacancy_salary = card.find('div', class_ = 'basic-salary').text
        vacancy_name = card.find('a', class_ = 'vacancy-card__title-link').text
        vacancy_addition = card.find('div', class_ = 'vacancy-card__meta')
        vacancy_skills = card.find('div', class_ ="vacancy-card__skills")
        company_name = card.find('a', class_ = 'link-comp link-comp--appearance-dark').text
        conditions = []
        skills = []
        for i in vacancy_addition.find_all('span', class_ = 'preserve-line'):
            conditions.append(i.text)
        for i in vacancy_skills.find_all('span', class_ = 'preserve-line'):
            skills.append(i.text)
        vacancies.append({'name': vacancy_name, 'salary': vacancy_salary, 'company': company_name, 'additional': conditions, 'skills': skills, 'date': vacancy_date})
```

```python
df = pd.DataFrame(vacancies)
df.to_csv('vacancies.csv')
```
